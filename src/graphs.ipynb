```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("../data/dataset_01.csv")
```

```python

vitoria = df["Vitoria"]
vitoria.value_counts(normalize=True).plot.pie(
    autopct='%1.1f%%',  # Formato da porcentagem (ex: 50.0%)
    startangle=90,  # Ângulo de rotação
    ylabel='',  # Remove o rótulo do eixo Y
    title='Porcentagem de vitória'
)

plt.show()
```

```python
df['N_de_tentativas'].value_counts().plot(kind='bar', color='skyblue', edgecolor='black')

# Adicionando título e rótulos para melhor visualização
plt.title('Frequência de Elementos Repetidos')
plt.xlabel('Número de tentativas')
plt.ylabel('Quantidade de vezes')

# Exibe o gráfico
plt.show()
```
